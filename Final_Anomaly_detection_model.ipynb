### 4. pre procesing

```python
anomaly_points = [
        ["2013-12-10 06:25:00.000000","2013-12-12 05:35:00.000000"],
        ["2013-12-15 17:50:00.000000","2013-12-17 17:00:00.000000"],
        ["2014-01-27 14:20:00.000000","2014-01-29 13:30:00.000000"],
        ["2014-02-07 14:55:00.000000","2014-02-09 14:05:00.000000"]
]
```

```python
data['timestamp']  = pd.to_datetime(data['timestamp'])

data['Anomaly'] = 0

for start, end in anomaly_points:
    data.loc[((data['timestamp']>=start) & (data['timestamp']<=end)),'Anomaly'] = 1
```

```python
#datetime information
```

```python
data['year'] = data['timestamp'].apply(lambda x : x.year)
data['month'] = data['timestamp'].apply(lambda x : x.month)
data['day'] = data['timestamp'].apply(lambda x : x.day)
data['hour'] = data['timestamp'].apply(lambda x : x.hour)
data['minute'] = data['timestamp'].apply(lambda x : x.minute)
```

```python
data.index = data['timestamp']
data.drop(['timestamp'], axis=1, inplace=True)
data.head(5)
```

### 6. Modeling(Iforest)

```python
iforest_model = IsolationForest(n_estimators=300, contamination=0.1, max_samples=700)
iforest_ret = iforest_model.fit_predict(data['value'].values.reshape(-1, 1))
iforest_df = pd.DataFrame()
iforest_df['value'] = data['value']
iforest_df['anomaly']  = [1 if i==-1 else 0 for i in iforest_ret]
```

```python
anomalies = [[ind, value] for ind, value in zip(iforest_df[iforest_df['anomaly']==1].index, iforest_df.loc[iforest_df['anomaly']==1,'value'])]
(hv.Curve(iforest_df['value'], label="Temperature") * hv.Points(anomalies, label="Detected Points").opts(color='red', legend_position='bottom', size=2, title="Isolation Forest - Detected Points"))\
    .opts(opts.Curve(xlabel="Time", ylabel="Temperature", width=700, height=400,tools=['hover'],show_grid=True))
```

```python
iforest_f1 = f1_score(data['Anomaly'], iforest_df['anomaly'])
print(f'Isolation Forest F1 Score : {iforest_f1}')
```

```python
## microsoft aomaly detector
```

```python
API_KEY = os.environ['ANOMALY_DETECTOR_API_KEY']
ENDPOINT = os.environ['ANOMALY_DETECTOR_ENDPOINT']
DATA_PATH = "./nab/realKnownCause/realKnownCause/machine_temperature_system_failure.csv"  # Example: c:\\test\\request-data.csv
CHUNK_SIZE = 5000  # Number of data points per request

client = AnomalyDetectorClient(ENDPOINT, AzureKeyCredential(API_KEY))

data_file = pd.read_csv(DATA_PATH)
timestamps = pd.to_datetime(data_file['timestamp'])

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(timestamps, data_file['value'], color='blue', label='Normal Data')

start_index = 0
total_points = len(data_file)
while start_index < total_points:
    end_index = min(start_index + CHUNK_SIZE, total_points)
    series = []
    for index, row in data_file.iloc[start_index:end_index].iterrows():
        series.append(TimeSeriesPoint(timestamp=row['timestamp'], value=row['value']))

    request = UnivariateDetectionOptions(series=series)
    change_point_response = client.detect_univariate_change_point(request)
    anomaly_response = client.detect_univariate_entire_series(request)

    for i, is_change_point in enumerate(change_point_response.is_change_point):
        if is_change_point:
            index = start_index + i
            ax.axvline(timestamps[index], color='orange', linestyle='--', linewidth=1, label='Change Point')

    for i, is_anomaly in enumerate(anomaly_response.is_anomaly):
        if is_anomaly:
            index = start_index + i
            ax.scatter(timestamps[index], data_file['value'][index], color='red', marker='o', label='Anomaly')

    start_index += CHUNK_SIZE

ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))
plt.xticks(rotation=45)
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title('Anomaly Detection')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import f1_score

# Assuming you have the true ground truth labels stored in the 'data' DataFrame column 'Anomaly'
ground_truth_labels = data['Anomaly'].astype(int)  # Convert to integer type (0 for normal, 1 for anomaly)

# Initialize an empty list to accumulate predicted labels
predicted_labels = []

for start_index in range(0, len(data_file), CHUNK_SIZE):
    end_index = min(start_index + CHUNK_SIZE, len(data_file))
    series = [TimeSeriesPoint(timestamp=row['timestamp'], value=row['value']) for index, row in data_file.iloc[start_index:end_index].iterrows()]

    request = UnivariateDetectionOptions(series=series)
    anomaly_response = client.detect_univariate_entire_series(request)

    # Accumulate the predicted labels
    predicted_labels.extend([1 if is_anomaly else 0 for is_anomaly in anomaly_response.is_anomaly])

# Calculate the F1 score
f1 = f1_score(ground_truth_labels, predicted_labels)

print("F1 Score:", f1)
```
